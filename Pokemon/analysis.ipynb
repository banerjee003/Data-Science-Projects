Most powerful pokemon from first 3 generation of type water

```python
df.loc[((df['Type 1'] == 'Water') | (df['Type 2'] == 'Water')) & (df['Generation'] <= 3)].sort_values(by='Total',ascending=False).head(5)
```

Most Powerful Dragon of last two generation

```python
df.loc[((df['Type 1'] == 'Dragon') | (df['Type 2'] == 'Dragon')) & (df['Generation'].isin ([5,6]))].sort_values(by='Total',ascending=False).head(5)
```

Select Most Powerful fire type pokemons

```python
powerful_fire = df.loc[(df['Type 1'] == 'Fire') | (df['Type 2'] == 'Fire')].sort_values(by='Total',ascending=False)
powerful_fire.head()
```
