9. Create a new column named performance based on exam score ranges.

```python
# code here
def grade(score):
    if score >= 45:
        return "Good"
    elif score >= 40:
        return "Average"
    else:
        return "Poor"
    
df['Performance'] = df['exam_score'].apply(grade)
df.head(10)
```
