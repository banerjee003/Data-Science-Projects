9. Create a new column named performance based on exam score ranges.

```python
# code here
def grade(score):
    if score >= 45:
        return "Good"
    elif score >= 40:
        return "Average"
    else:
        return "Poor"
    
df['Performance'] = df['exam_score'].apply(grade)
df.head(10)
```

15. Create study_sleep_ratio column and analyze its impact.

```python
# code here
df['study_sleep_ratio'] = df['study_hours'] / df['sleep_hours']
df.head()
```

```python
sns.scatterplot(x='study_sleep_ratio', y='exam_score', data=df)
plt.title("Study Sleep Ratio vs Exam Score")
plt.show()
```

## Insights from the Analysis

1. **Attendance is a key factor in performance**
   Students with higher attendance consistently achieved better exam scores. Those in the 90–100% attendance range performed significantly better than students below 70%, showing that regular class participation strongly influences results.

2. **Study hours alone do not guarantee high marks**
   Although there is a positive relationship between study hours and exam score, the pattern is not perfectly linear. Some students with high study hours still scored low, indicating that study quality and other factors also matter.

3. **Balanced study and sleep leads to better outcomes**
   The study_sleep_ratio analysis shows that students with a balanced ratio between study time and sleep performed better than those who studied excessively with less sleep. Lack of proper rest negatively impacted scores.

4. **Previous academic performance affects current results**
   Students with higher previous grades generally scored better in the current exam, suggesting that strong academic foundations and consistent learning over time play an important role.

5. **Internet usage has minimal direct impact**
   Internet_hours showed very weak correlation with exam_score, indicating that the amount of online time alone does not determine performance. How the internet is used is likely more important than total hours.
