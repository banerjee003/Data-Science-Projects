```python
df['highest wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest wickets'] = df['highest wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest wickets'].sum().sort_values(ascending=False)
final = top_bowlers.head(10)
```

```python
sns.barplot(x=final,y=final.index)
plt.title('Highest Wicket Takers')
```
